```python
#import CSV

import pandas as pd
dr = pd.read_csv(r'C:\drucker.csv',encoding='cp1252')

#print the first few rows
dr.head()
```

```python
#remove punctuation

df = pd.DataFrame(dr)

df["Unified_Name"] = df['COMPANY'].str.replace('[^\w\s]','')

df.head()
```

```python
#tokenize dataframe

import nltk
nltk.download('punkt')

from nltk import word_tokenize

df['Unified_Name'].apply(word_tokenize)
```

```python
#strip a list of words from dataframe

stopwords = ['Company','Corp','Inc','Co','Ltd','plc','PLC','Group','Grp','Holdings','A','B','C']

df['Unified_Name'] = df['Unified_Name'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopwords)]))
```

```python
#export dataframe to CSV

df.to_csv('drucker_new.csv')
```
